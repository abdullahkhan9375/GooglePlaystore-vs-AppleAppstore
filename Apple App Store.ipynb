```python
import matplotlib.pyplot as plt
import collections as cl
import numpy as np
import seaborn as sns
import pandas as pd
import plotly.express as px
```

```python
path1 = "datasets/AppleStore.csv"
dat1 = pd.read_csv(path1)
dfApple = pd.DataFrame(dat1)
dfApple.head()
```

```python
dfApple = dfApple.drop(columns = ["currency", "id", "vpp_lic", "ver", "Unnamed: 0","lang.num", "user_rating_ver","rating_count_ver", "sup_devices.num","ipadSc_urls.num"])
```

```python
dfApple.head()
```

```python
#data cleaning
#free or not
dfApple = dfApple.rename(columns = {"rating_count_tot": "Reviews", "user_rating" : "Rating", "cont_rating" : "Content Rating",
                          "prime_genre" : "Category", "price" : "Price", "size_bytes" : "Size", "track_name" : "App"})
dfApple["Type"] = "1"
```

```python
dfApple["Category"].unique()
```

```python
dfApple[dfApple['Category'] == 'Games']
```

```python
dfApplefree = dfApple[dfApple["Price"] == 0] 
dfApplepaid = dfApple[dfApple["Price"] > 0]
dfApplepaid["Type"] = "1"
dfApplefree["Type"] = "0"
dfAppleclean = pd.concat([dfApplefree, dfApplepaid])
dfAppleclean["Category"].unique()
```

```python
#clean content rating
def cleanContent(x):
    if x == "4+":
        return "Everyone"
    elif x == "9+":
        return "Everyone 10+"
    elif x == "12+":
        return 'Teen'
    elif x == "17+":
        return 'Mature 17+'

def cleanCategory(x):
    if x == "Reference" or x == "Book":
        return "Books & Reference"
    else:
        return x
```

```python
dfApple = dfAppleclean
dfApple["Content Rating"] = dfApple["Content Rating"].replace([x for x in dfApple["Content Rating"]], [cleanContent(x) for x in dfApple["Content Rating"]])
dfApple["Category"] = dfApple["Category"].replace([x for x in dfApple["Category"]], [cleanCategory(x) for x in dfApple["Category"]])
```
